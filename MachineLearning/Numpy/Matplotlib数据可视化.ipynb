```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
```

```python
cosy.shape
```

```python
n = 12 # 生成n个数据
X = np.arange(n) # X为0-1的整数
Y1 = (1 - X / float(n)) * np.random.uniform(0.5,1.0,n)# 随机分布
Y2 = (1 - X / float(n)) * np.random.uniform(0.5,1.0,n)# 随机分布
```

```python
plt.bar(X,+Y1) # 华仔X轴的上方
plt.bar(X,-Y2) # 华仔X轴的下方
# 设置横纵坐标的边界，去掉横坐标的线
plt.xlim(-.5, n)
plt.xticks(())
plt.ylim(-1.25, 1.25)
plt.yticks(())

# 加颜色优化图像，facecolor设置主体颜色，edgecolor设置边框颜色
plt.bar(X,+Y1,facecolor="#9999ff",edgecolor="white")
plt.bar(X,-Y2,facecolor="#ff9999",edgecolor="white")

for x,y in zip(X,Y1):
    plt.text(x,y + 0.05,'%.2f' % y,ha='center',va='bottom')
for x,y in zip(X,Y2):
    plt.text(x,-y - 0.15,'%.2f' % y,ha='center',va='bottom')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D #的 导入3D图形库
```

```python
fig = plt.figure() # 先定义一个图像窗口
ax = Axes3D(fig) # 在窗口上添加3D坐标轴
```

```python
fig = plt.figure() # 先定义一个图像窗口
ax = Axes3D(fig) # 在窗口上添加3D坐标轴
# 定义数据
X = np.arange(-4,4,0.1) # 随机数据
Y = np.arange(-4,4,0.1) # s随即数据
X,Y = np.meshgrid(X,Y) # x-y 平面的网络
R = np.sqrt(X ** 2 + Y ** 2)
# 计算高度值
Z = np.sin(R)

 # 填充颜色
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow')) # rstride:row的跨度 cstride: col的跨度

# 添加XY平面的等高线
ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap=plt.get_cmap('rainbow')) # zdir:沿着z轴方向投影，offset相当于z=0的偏移距离

ax.set_zlim(-2,2) # 设置z轴的坐标范围
plt.show()
```
