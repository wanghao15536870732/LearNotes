### scilit-learn中的Scaler

```python
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris() # 加载鸢尾花的数据
```

```python
X = iris.data # 记载测试数据
y = iris.target # 记载标签
```

```python
X[:10,:] # 查看前10行的内容
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_Train,X_test,y_train,y_test = train_test_split(iris.data,iris.target,random_state=666) # 分成训练数据集跟测试数据集
```

### scilit-learn中的StandardScaler  均值方差归一化

```python
from sklearn.preprocessing import StandardScaler # 数据预处理包中import
```

```python
standardScaler = StandardScaler()
standardScaler.fit(X_Train) # 将返回自身关键信息
```

```python
standardScaler.mean_ # 四个特征对应的均值
```

```python
standardScaler.scale_ # 描述数据的分布范围,即四个特征对应的方差
```

```python
standardScaler.transform(X_Train) # 虽数据进行归一化处理
```

```python
X_Train = standardScaler.transform(X_Train) # 改变一下x_train
```

```python
X_Train
```

```python
standardScaler.transform(X_test)
```

```python
X_test_standerd = standardScaler.transform(X_test) # 保存归一化后的数据
```

```python
X_test_standerd
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=3) # 创建分类器
```

```python
knn_clf.fit(X_Train,y_train)
```

```python
knn_clf.score(X_test_standerd,y_test) # 此时分类器的准确度
```

```python
knn_clf.score(X_test,y_test) # 取消数据归一化后分类器的准确度
```
