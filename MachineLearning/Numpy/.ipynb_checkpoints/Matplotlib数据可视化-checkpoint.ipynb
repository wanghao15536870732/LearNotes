```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y) # 横纵坐标对应的参数，返回一个对象 绘制曲线
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red") # 改变颜色
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--") # 改变线得风格
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--") # 改变线得风格
plt.xlim(-5,15) # 改变x坐标得范围
plt.ylim(0,1.5) # 改变y坐标得范围
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--") # 改变线得风格
plt.axis([-1,11,-2,2]) # 同时改变xy轴得范围 x轴范围-1 - 11  x轴范围-2 - 2
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--") # 改变线得风格
plt.xlabel("x axis") # x轴线得文字输出
plt.ylabel("y value") # y轴线得文字输出
plt.show()
```

```python
plt.plot(x,siny,label="sin(x)") # 通过label添加图示
plt.plot(x,cosy,color="red",linestyle="--",label="cos(x)")  # 通过label添加图示
plt.xlabel("x axis") # x轴线得文字输出
plt.ylabel("y value") # y轴线得文字输出
plt.legend() # 渲染图示
plt.show()
```

```python
plt.plot(x,siny,label="sin(x)") # 通过label添加图示
plt.plot(x,cosy,color="red",linestyle="--",label="cos(x)")  # 通过label添加图示
plt.xlabel("x axis") # x轴线得文字输出
plt.ylabel("y value") # y轴线得文字输出
plt.legend() # 渲染图示
plt.title("Welcome to the ML World!!") # 添加标题
plt.show()
```

### Scatter Plot  散点图

```python
plt.scatter(x,siny) # 绘制散点图
plt.show()
```

```python
plt.scatter(x,siny) # 绘制散点图
plt.scatter(x,cosy,color="red")
plt.show()
```

```python
x = np.random.normal(0,1,100) # 随机生成0-1之间100个正态分布得数
y = np.random.normal(0,1,100) 

plt.scatter(x,y)
plt.show()
```

```python
x = np.random.normal(0,1,10000) # 随机生成0-1之间100个正态分布得数
y = np.random.normal(0,1,10000) 

plt.scatter(x,y)
plt.show()
```

```python
x = np.random.normal(0,1,10000) # 随机生成0-1之间100个正态分布得数
y = np.random.normal(0,1,10000) 

plt.scatter(x,y,alpha=0.3) # alpha参数改变透明度
plt.show()
```
