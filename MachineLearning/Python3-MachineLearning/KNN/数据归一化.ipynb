### 数据归一化处理

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 最值归一化 Normalization

```python
x = np.random.randint(0,100,size=100)
```

```python
x
```

```python
(x - np.min(x))  / (np.max(x) - np.min(x))
```

```python
X = np.random.randint(0,100,(50,2)) # 只有整型
```

```python
X[:10,:]
```

```python
X = np.array(X,dtype=float)
```

```python
X[:10,:]
```

```python
X[:,0] = (X[:,0] - np.min(X[:,0])) / (np.max(X[:,0] - np.min(X[:,1]))) # 对第一列最值归一化
```

```python
X[:,1] = (X[:,1] - np.min(X[:,1])) / (np.max(X[:,1] - np.min(X[:,1]))) # 对第二列最值归一化
```

```python
X[:10,:]
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
np.mean(X[:,0]) # 第0列对应的均值
```

```python
np.std(X[:,0]) # 第0列对应的方差
```

```python
np.mean(X[:,1]) # 第1列对应的均值
```

```python
np.std(X[:,1]) # 第1列对应的方差
```

### 均值方差归一化 Standerdization

```python
X2 = np.random.randint(0,100,(50,2))
```

```python
X2 = np.array(X2,dtype=float)
```

```python
X2[:,0] = (X2[:,0] - np.mean(X2[:,0])) / np.std(X2[:,0])
```

```python
X2[:,1] = (X2[:,1] - np.mean(X2[:,1])) / np.std(X2[:,1])
```

```python
plt.scatter(X2[:,0],X2[:,1])
plt.show()
```

```python
np.mean(X2[:,0]) # 第0列的均值
```

```python
np.std(X2[:,0]) # 第0列的方差
```

```python
np.mean(X2[:,1]) # 第1列的均值
```

```python
np.std(X2[:,1]) # 第1列的方差
```
